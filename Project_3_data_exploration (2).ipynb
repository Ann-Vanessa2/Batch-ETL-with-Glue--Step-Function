### Extracting Data from Redshift and Storing in S3

Testing locally

```python
import pymysql
import pandas as pd
import boto3
import os
from google.colab import userdata
from io import StringIO
```

```python
# Retrieve AWS credentials
AWS_ACCESS_KEY = userdata.get('AWS_ACCESS_KEY_ID')
AWS_SECRET_KEY = userdata.get('AWS_SECRET_ACCESS_KEY')

# Set as environment variables for Boto3
os.environ["AWS_ACCESS_KEY_ID"] = AWS_ACCESS_KEY
os.environ["AWS_SECRET_ACCESS_KEY"] = AWS_SECRET_KEY
```

```python
# Retrieve RDS credentials
RDS_HOST = userdata.get("RDS_HOST")
RDS_USER = userdata.get("RDS_USER")
RDS_PASSWORD = userdata.get("RDS_PASSWORD")
RDS_DB = "rental_marketplace"
S3_BUCKET = "rental-marketplace"
```

```python
# Connect to RDS
conn = pymysql.connect(
    host=RDS_HOST, user=RDS_USER, password=RDS_PASSWORD, database=RDS_DB
)

# Fetch Data Function
def fetch_and_upload(table_name):
    query = f"SELECT * FROM {table_name}"
    df = pd.read_sql(query, conn)

    # Convert to CSV and upload to S3
    csv_buffer = StringIO()
    df.to_csv(csv_buffer, index=False)
    s3_client = boto3.client("s3")
    s3_client.put_object(
        Bucket=S3_BUCKET,
        Key=f"raw-data/{table_name}.csv",
        Body=csv_buffer.getvalue(),
    )
    print(f"{table_name} uploaded to S3")

# Extract all tables
tables = ["apartment_attributes", "apartments", "bookings", "user_viewing"]
for table in tables:
    fetch_and_upload(table)

# Close connection
conn.close()
```
